Importing things to use nltk & wordcloud:

```python
import nltk
from nltk.corpus import stopwords, words

# these should be "native" to Python
import string
import itertools # might have to install?

from wordcloud import WordCloud

import matplotlib.pyplot as plt

%matplotlib inline
```

Let's break this into individual words:

```python
myWords = text.split()
```

```python
myWords
```

One simple thing we can think of doing is making a bar plot showing the frequency of words.

```python
wordsDict = {} # start with empty dictionary
for word in myWords: # loop over all words in text
    if word in wordsDict: # if already there
        wordsDict[word] += 1 # increase the count of this word
    else: # then add it to the dictionary
        wordsDict[word] = 1
```

```python
len(wordsDict)
```

Let's put this in a format that will show the 50 (or another number) most frequent words in our text.

```python
for k,v in wordsDict.items():
    print(k, v)
```

```python
import operator # to help us sort our dict by value
wordsDict = sorted(wordsDict.items(), key=operator.itemgetter(1)) # sort dict by 2nd component (i.e. value)
```

```python
wordsDict[::-1]
```

This returns as a list, let's put it back into a dictionary:

```python
wordsDict = dict(wordsDict[::-1])
```

```python
wordsDict
```

Let's actually plot the 50 most frequent words here:

```python
n = 50
wordsPlot = {k: wordsDict[k] for k in list(wordsDict)[:n]} # sorting magic, because we can't index first nth of dict
wordsPlot
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

ax.set_xticklabels(list(wordsPlot.keys()))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

ax.set_xticks(range(len(wordsPlot))) # add in the correct ticks at the right bars
ax.set_xticklabels(list(wordsPlot.keys())) # adding in words

plt.show()
```

We want to "clean" our text dataset of these frequent words in the English language -- aka "stopwords".

```python
# from the nltk package
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
less_words = {} # this will be our text, but devode of "stop words"
for w, num in wordsDict.items():
    if w.lower() not in stop_words:
        less_words[w] = num
```

```python
less_words
```

```python
n = 50
wordsPlot = {k: less_words[k] for k in list(less_words)[:n]} # sorting magic, because we can't index first nth of dict
#wordsPlot

fig, ax = plt.subplots(figsize=(15,3))
ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

ax.set_xticks(range(len(wordsPlot))) # add in the correct ticks at the right bars
ax.set_xticklabels(list(wordsPlot.keys())) # adding in words

# rotate our tick labels
plt.xticks(rotation=90)

plt.show()
```

We want to take out punctuation, but this can be a little tricky.

"I'll" is different than "Ill" -- first is contraction of "I will" and the seconds means "sick".

There is also captialization to worry about. "Why did you..." vs. "That is why I did..." -- here we'd want to "count" this "why" as the same thing.  We're ignore this for now.

All we are going to do is take out trailing punctuation.  So that: "his," and "his" are noted as the same word.

```python
string.punctuation
```

```python
no_punc_words = {}
for word, value in less_words.items():
    #print(word, value)
    if word[-1] in string.punctuation: # if the last element of the word is a punctuation
        word = word[:-1] # this takes out last punctuation mark
    if len(word) > 0: # taking out *only* punctuation marks
        if word in no_punc_words: # if already there
            no_punc_words[word] += value # increase the count of this word
        else: # then add it to the dictionary
            no_punc_words[word] = value
```

```python
# re-sort by 50 most frequent
no_punc_sorted = sorted(no_punc_words.items(), key=operator.itemgetter(1)) # sort dict by 2nd component (i.e. value)
no_punc_sorted = dict(no_punc_sorted[::-1])
```

```python
n = 50
wordsPlot = {k: no_punc_sorted[k] for k in list(no_punc_sorted)[:n]} # sorting magic, because we can't index first nth of dict
#wordsPlot

fig, ax = plt.subplots(figsize=(15,3))
ax.bar(range(len(wordsPlot)), list(wordsPlot.values()), align='center')

ax.set_xticks(range(len(wordsPlot))) # add in the correct ticks at the right bars
ax.set_xticklabels(list(wordsPlot.keys())) # adding in words

# rotate our tick labels
plt.xticks(rotation=90)

plt.show()
```

After all of that formatting, the wordcloud creation is relatively easy:

```python
no_punc_sorted
```

```python
wordcloud = WordCloud().generate_from_frequencies(no_punc_sorted)
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.imshow(wordcloud)

plt.show()
```
